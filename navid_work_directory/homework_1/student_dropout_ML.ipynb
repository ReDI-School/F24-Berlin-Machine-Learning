```python
## convert categorial data to numerical or boolean
df['Gender'] = df['Gender'].map({'F': 1, 'M': 0})
df['School'] = df['School'].map({'GP': 0, 'MS': 1})
df['Address'] = df['Address'].map({'U': 0, 'R': 1})
df['Parental_Status'] = df['Parental_Status'].map({'T': 0, 'A': 1})
df['Family_Size'] = df['Family_Size'].map({'GT3': 0, 'LE3': 1})
df['Guardian'] = df['Guardian'].map({'father': 0, 'mother': 1, 'other':2})
# df['Mother_Job'] = df['Mother_Job'].map({'at_home': 0, 'services': 1, 'health':2, 'teacher': 3, 'other': 4})
# df['Father_Job'] = df['Father_Job'].map({'at_home': 0, 'services': 1, 'health':2, 'teacher': 3, 'other': 4})
# df['Reason_for_Choosing_School'] = df['Reason_for_Choosing_School'].map({'course': 0, 'home': 1, 'reputation':2, 'other':3})


df['School_Support'] = df['School_Support'].map({'no': False, 'yes': True}).astype('bool')
df['Family_Support'] = df['Family_Support'].map({'no': False, 'yes': True}).astype('bool')
df['Extra_Paid_Class'] = df['Extra_Paid_Class'].map({'no': False, 'yes': True}).astype('bool')
df['Extra_Curricular_Activities'] = df['Extra_Curricular_Activities'].map({'no': False, 'yes': True}).astype('bool')
df['Attended_Nursery'] = df['Attended_Nursery'].map({'no': False, 'yes': True}).astype('bool')
df['Wants_Higher_Education'] = df['Wants_Higher_Education'].map({'no': False, 'yes': True}).astype('bool')
df['Internet_Access'] = df['Internet_Access'].map({'no': False, 'yes': True}).astype('bool')
df['In_Relationship'] = df['In_Relationship'].map({'no': False, 'yes': True}).astype('bool')
```

```python
#Dropping some columns
df = df.drop("Mother_Job", axis=1)
df = df.drop("Father_Job", axis=1)
df = df.drop("Reason_for_Choosing_School", axis=1)
```

```python
#Create training and testing dataset
from sklearn.model_selection import train_test_split
X = df.drop('Dropped_Out', axis=1) 
y = df['Dropped_Out']               # Target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# Train the Model: Fit the model to the training data.
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
#Test the model and evaluate its performance.
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%') #Check accuracy of model
```
