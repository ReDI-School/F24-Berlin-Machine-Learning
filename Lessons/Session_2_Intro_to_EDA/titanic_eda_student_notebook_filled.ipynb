Now we need to have a look at what data is going to be useful going forward for our analysis, something like name doesnt really matter too much so we can drop it. Wh else could we drop? We will also need to look at converting our categorical data into numerical. In the coming lessons we will show you about using OneHotEncoding and get dummies which create a column for each available option in a column, for example it would split gender into 2 columns, male and female and assign a 1 or 0 to this, in fact we only need one column for this, we can just have 1 or 0 for either male or female because the alternative is the other. 

```python
df1= df.drop(['Name'], axis=1)
df1['Gender'] = df1['Gender'].map({'male': 0, 'female': 1})

df1.head()
```

Lets also convert the age to an int64

```python
df1['Age'] = df['Age'].astype('int64')
df1.head()
```

Lets drop the Ticket and Cabin number for now, currently it doesnt seem useful, you can always add it back later

```python
df1= df1.drop(['Ticket', 'Cabin'], axis=1)
df1.head()
```

we still have the Emarked feature so lets map that to 1,2,3 just like we did with the gender and lets make it more easy to understand

```python
df1['Embarked_from'] = df1['Embarked'].map({'C': 0, 'Q': 1, 'S':2})
df1.head()
```

now we also need to drop or old embarked column

```python
df1= df1.drop(['Embarked'], axis=1)
df1.head()
```
